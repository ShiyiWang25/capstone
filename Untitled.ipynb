```python
import numpy as np
 
def detect_outlier(data):
    # find q1 and q3 values
    q1, q3 = np.percentile(sorted(data), [25, 75])
 
    # compute IRQ
    iqr = q3 - q1
 
    # find lower and upper bounds
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
 
    outliers = [x for x in data if x <= lower_bound or x >= upper_bound]
 
    return outliers
 
# input data
test_3 = [7.18118190201366,6.81372816790546,8.17866935834271,7.62574092542886,5.74176708476276,8.83352550069852,9.67070362937981,7.48361411588081,5.96663314385443,8.20100011608628,5.82863150277497,5.52036142289113,6.70795925617069,7.20140173446171,8.31765328013218,5.82575650263386,7.600232381781,5.83920506615225,8.60386405160954,8.80976828561479,5.30447799729141,10.1177298307827,9.19280772243909,7.98831903244424,6.32042018144661,7.67194699614761,7.96117413481845,8.03707896953035,8.11842988080128,8.97044729795703,7.72278080079145,8.77878675241107,9.41098514057407,6.02582332587479,7.28659617087145,10.2029154992826,6.13469086103084,3.94935671134126,6.11401033058138,]
detect_outlier(test_3)
```
